### 1.1 What is Deep Learning?

Deep learning is a subfield of machine learning that focuses on algorithms inspired by the structure and function of the human brain, called artificial neural networks. These algorithms enable machines to learn complex patterns and representations from large amounts of data. Deep learning has led to significant advancements in various domains, including computer vision, natural language processing, and speech recognition.

### 1.2 The difference between Artificial Intelligence, Machine Learning, and Deep Learning

* **Artificial Intelligence (AI)** is the broader concept of creating machines or systems that can perform tasks that would typically require human intelligence, such as problem-solving, understanding language, and recognizing patterns.
* **Machine Learning (ML)** is a subset of AI that focuses on developing algorithms that enable machines to learn from and make predictions or decisions based on data without being explicitly programmed.
* **Deep Learning (DL)** is a subfield of ML that specifically utilizes artificial neural networks to model and solve complex problems. It can automatically learn hierarchical feature representations from raw data, making it particularly effective for tasks involving unstructured data.

## 2 What are Artificial Neural Networks (ANNs)?

Artificial Neural Networks (ANNs) are computational models inspired by the biological neural networks of the human brain. They are designed to mimic the way the brain processes information and learns from it. ANNs consist of interconnected nodes, called neurons or units, which are organized into layers. They are used to recognize patterns, classify data, and make predictions, making them applicable in various fields such as computer vision, natural language processing, and speech recognition.

###  Basic Concepts

* **Neurons** : The basic building blocks of ANNs, neurons receive input from other neurons or external data sources, perform a computation, and generate an output. Each neuron is connected to other neurons through weighted connections.
* **Layers**: Neurons are organized into layers. There are three types of layers in ANNs: input layer (receives raw input data), hidden layers (process the input and produce an output), and output layer (produces the final output or prediction).
* **Weights**: Weights are numerical values associated with the connections between neurons. They determine the strength of the connection and are adjusted during the training process to minimize the error between the predicted and actual outputs.
* **Biases**: Biases are additional constants added to the weighted sum of inputs in a neuron. They allow for better flexibility and control over the neuron's output, acting as a shift in the activation function.
* **Activation Functions** : Activation functions are mathematical functions applied to the output of a neuron. They introduce non-linearity into the network, enabling it to learn complex relationships between inputs and outputs. Common activation functions include ReLU, sigmoid, and tanh.

### Types of ANNs

* Feedforward Neural Networks (FNN): In FNNs, information flows in a single direction from the input layer to the output layer without looping back. They are the simplest type of ANN and are commonly used for tasks such as regression and classification.


* Recurrent Neural Networks (RNN): RNNs have connections that loop back, allowing them to maintain a hidden state that can capture information from previous time steps. This makes them suitable for processing sequences of data, such as time series and natural language.


* Convolutional Neural Networks (CNN): CNNs are designed to process grid-like data, such as images. They use convolutional layers to scan the input for local patterns, enabling them to learn hierarchical features and achieve robust performance in tasks like image classification and object detection.

    
Other specialized ANNs include Long Short-Term Memory (LSTM) networks, Gated Recurrent Units (GRUs), and Transformer models, which have their specific applications and characteristics.

### Keras Sequential and Functional API

Keras provides two main ways to define neural network models: the Sequential API and the Functional API.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(32, input_shape=(784,), activation='relu'))
model.add(Dense(10, activation='softmax'))
```

* **Functional API**: A more flexible way to define models with complex architectures, such as multiple inputs/outputs, shared layers, or non-linear topologies. You create a functional model by defining input and output tensors and specifying their relationships.

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

inputs = Input(shape=(784,))
x = Dense(32, activation='relu')(inputs)
outputs = Dense(10, activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)
```

#### Loss functions

Loss functions, also known as cost functions or objective functions, quantify the difference between the predicted output and the actual output (ground truth) for a particular input in a machine learning model. They are used to guide the optimization process and help the model learn from the training data. Here are some common loss functions and their typical use cases:



* **Mean Squared Error (MSE)**: MSE measures the average squared difference between the predicted values and the actual values. It is widely used in regression tasks and is sensitive to outliers due to the squaring term.

     * Usage: Regression tasks, linear regression, feedforward neural networks for regression
     * Formula: MSE = (1/n) * Σ(actual - predicted)^2


* **Mean Absolute Error (MAE)**: MAE measures the average absolute difference between the predicted values and the actual values. It is also used in regression tasks and is less sensitive to outliers compared to MSE.

    * Usage: Regression tasks, robust regression, feedforward neural networks for regression
    * Formula: MAE = (1/n) * Σ|actual - predicted|



* **Cross-Entropy Loss (Log Loss)**: Cross-entropy loss measures the dissimilarity between two probability distributions – the predicted probabilities and the true probabilities (one-hot encoded) for classification tasks. It is the most commonly used loss function for classification problems with multiple classes.

    * Usage: Multi-class classification, logistic regression, feedforward neural networks for classification
    * Formula: CrossEntropy = -Σ(y * log(y_pred)), where y is the true probability distribution, and y_pred is the predicted  probability distribution



* **Binary Cross-Entropy Loss**: Binary cross-entropy loss is a special case of cross-entropy loss for binary classification problems. It measures the dissimilarity between the predicted probability and the actual label for each class.

    * Usage: Binary classification, logistic regression, feedforward neural networks for binary classification
    * Formula: BinaryCrossEntropy = -Σ(y * log(y_pred) + (1-y) * log(1-y_pred)), where y is the true label, and y_pred is the predicted probability


* **Hinge Loss**: Hinge loss is used for training Support Vector Machines (SVM) in binary classification tasks. It measures the error between the predicted margin and the actual margin.

   * Usage: Support Vector Machines, maximum margin classifiers
   * Formula: HingeLoss = max(0, 1 - y * y_pred), where y is the true label (-1 or 1), and y_pred is the predicted margin


* **Huber Loss**: Huber loss is a combination of MSE and MAE. It is less sensitive to outliers than MSE and smoother than MAE. It measures the squared error for small errors and the absolute error for larger errors, with a threshold (δ) to differentiate between the two.

    * Usage: Regression tasks, robust regression, feedforward neural networks for regression
    * Formula: HuberLoss = 0.5 * (actual - predicted)^2 if |actual - predicted| <= δ, else δ * |actual - predicted| - 0.5 * δ^2

### Overfitting and Underfitting: Strategies to Prevent Them

Overfitting occurs when the model learns the training data too well, including its noise, resulting in poor generalization to new data. Underfitting occurs when the model fails to capture the underlying structure of the data, resulting in low performance on both the training and test data.


Strategies to prevent overfitting and underfitting include:

* Regularization (L1, L2, or Dropout)
* Early stopping
* Data augmentation
* Using a simpler or more complex architecture (depending on whether underfitting or overfitting is the issue)

The main difference between L1 and L2 regularization is that L1 regularization encourages sparsity in the model weights, while L2 regularization does not. Sparsity means that many of the weights are zero, which can make the model more interpretable.

* F1 Score: The harmonic mean of precision and recall, providing a balanced measure of the model's performance.

## 7 Tuning Neural Networks

### Regularization Techniques: L1, L2, and Dropout

Regularization techniques are used to prevent overfitting by adding a penalty to the loss function based on the complexity of the model.

* **L1 Regularization**: Adds the absolute value of the weights multiplied by a regularization constant (lambda) to the loss function. L1 regularization promotes sparsity in the model weights.
* **L2 Regularization**: Adds the square of the weights multiplied by a regularization constant (lambda) to the loss function. L2 regularization encourages small weights but does not enforce sparsity.
* **Dropout**: Randomly drops a certain percentage of neurons during training, preventing the model from becoming overly reliant on any single neuron.

```python
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import L1, L2

model.add(Dense(64, activation='relu', kernel_regularizer=L1(0.01), bias_regularizer=L1(0.01)))
model.add(Dense(64, activation='relu', kernel_regularizer=L2(0.01), bias_regularizer=L2(0.01)))
model.add(Dropout(0.5))
```

### Early Stopping and Learning Rate Scheduling

Early Stopping: Monitors a specified metric (e.g., validation loss) during training and stops training when the metric stops improving for a certain number of epochs. This prevents overfitting and reduces the time spent on training.

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_val, y_val), callbacks=[early_stopping])
```

### Learning Rate Scheduling:
    
Adjusts the learning rate during training according to a predefined schedule or based on the model's performance. This can help the model converge faster and find better minima in the loss function.

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=0.0001)
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_val, y_val), callbacks=[reduce_lr])
```

Combining these techniques, such as early stopping and learning rate scheduling, can help improve the model's performance and efficiency.

### Autoencoders for Dimensionality Reduction and Denoising

Autoencoders are unsupervised neural networks that learn to compress and reconstruct input data. 

* They consist of two parts:

## 9 Transfer Learning

#### What is Transfer Learning?

Transfer learning is a technique in machine learning where a pre-trained model is used as a starting point for training a new model on a related task. The idea is to leverage the knowledge gained from solving one problem and applying it to a different but related problem. This approach often leads to faster training and better performance, especially when the new task has limited labeled data.

### Pre-trained Models in Keras

Keras provides several pre-trained models for image classification tasks, trained on large datasets like ImageNet. These models include:

* VGG16
* VGG19
* ResNet50
* InceptionV3
* Xception
* MobileNet

You can import a pre-trained model using the tensorflow.keras.applications module:

```python
from tensorflow.keras.applications import VGG16

base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

* The weights parameter specifies the pre-trained weights to use, include_top indicates whether to include the fully connected layer at the top of the model, and input_shape defines the input tensor shape.

###  Fine-tuning and Feature Extraction

There are two common approaches to transfer learning: fine-tuning and feature extraction.
    
    
* **Fine-tuning**: In this approach, you replace the top layers of the pre-trained model with new layers specific to your task and train these new layers along with a few lower layers of the model. This enables the model to learn features specific to your task while retaining the knowledge learned from the original task.

* **Feature Extraction**: In this approach, you use the pre-trained model as a fixed feature extractor by removing the top layers and running the remaining model on your dataset. The output features are then used as input to a new, smaller model trained specifically for your task.

## 10 Simple feedforward neural network using Keras

we'll build a simple feedforward neural network using Keras to classify handwritten digits from the MNIST dataset

### Load and preprocess the data

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize pixel values to the range [0, 1]
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255

# One-hot encode labels
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# Flatten images to be input into ANN
train_images = train_images.reshape(train_images.shape[0], -1)
test_images = test_images.reshape(test_images.shape[0], -1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Flatten images
train_images_flattened = train_images.reshape(train_images.shape[0], -1)

# Choose a random image index
random_index = np.random.randint(0, train_images_flattened.shape[0])

# Get the flattened image and reshape it to (28, 28)
image = train_images_flattened[random_index].reshape(28, 28)

# Display the image and its corresponding label
plt.imshow(image, cmap='gray')
plt.title(f"Label: {train_labels[random_index]}")
plt.axis('off')
plt.show()
```

### Build the neural network model:

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

### Compile the model:

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

### Train the model:

```python
history = model.fit(train_images, train_labels, epochs=10, batch_size=128, validation_split=0.1)
```

### Evaluate the model:

```python
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print(f"Test accuracy: {test_accuracy:.4f}")
```

### Make predictions:

```python
predictions = model.predict(test_images)
predicted_labels = np.argmax(predictions, axis=1)
```
