```python
#read csv in a folder 
df=pd.read_csv("./Sales_Data/Sales_April_2019.csv")
df.head()
```

```python
#How do I list or read all files of a directory?

import os
# help us list file in folders 
files =[file for file in os.listdir('./Sales_Data')]

for file in files:
    print (file)
```

```python


#empty data frame that will contain 12 month data AFTERWARD 
all_months_data= pd.DataFrame()

files =[file for file in os.listdir('./Sales_Data')]

for file in files:
    
    df=pd.read_csv("./Sales_Data/"+file)
    # Concatenating multiple csvs together to create a new DataFrame (pd.concat) 
    all_months_data=pd.concat([all_months_data, df])

all_months_data.head()    
```

```python
#save all months data as csv
all_months_data.to_csv("all_data.csv",index=False)
```

```python
#read the new csv containing all data 
all_data=pd.read_csv("all_data.csv")
all_data.head()
```

drop rows of NAN

```python
nan_df=all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

# Clean up the data 

```python
all_data=all_data.dropna(how='all')
all_data.head()
```

```python
#find or and delete it'
temp_df=all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df.head()
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

convert columns to correct type 

```python
#make int
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])

#make float 
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
```

Augment the data with addtional column

add month column

```python
all_data['Month']= all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')
all_data.head()
```

#add sales column

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

# 1)What was the best month for sales? How much was earned that month?

```python
#TOTAL SALES (REVENUE 2019)
all_data['Sales'].sum()
```

```python
all_data.groupby('Month').sum()
```

```python
results=all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
months =range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('sales in usd ')
plt.xlabel('Month number ')
plt.show()
```

# What city sold the most product?

add city column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

#all_data['City']=all_data['Purchase Address'].apply(lambda x:get_city(x) + ' ' + get_state(x))
all_data['City']=all_data['Purchase Address'].apply(lambda x:f"{get_city(x)} ({get_state(x)})")
#all_data['City']=all_data['Purchase Address'].apply(lambda x:x.split(',')[1])
#all_data=all_data.drop(columns="column")
all_data.head()
```

```python
results=all_data.groupby('City').sum()
```

```python
results
```

```python
import matplotlib.pyplot as plt
#cities =all_data['City'].unique()

cities= [city  for city, df in all_data.groupby('City')]

plt.bar(cities,results['Sales'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel('sales in usd ')
plt.xlabel('Month number ')
plt.show()
```

# What time should we display advertisemens to maximize the likelihood of customerâ€™s buying product?

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute']=all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours=[ hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
#all_data.groupby(["Hour"]).count()
```

# What products are most often sold together?

```python
all_data.head()
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head()
```

```python
df=df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count =Counter()

for row in df['Grouped']:
    row_list =row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

# What product sold the most? Why do you think it sold the most?

```python
product_group=all_data.groupby('Product')
```

```python
print(product_group.sum())
```

```python
quantity_ordered=product_group.sum()['Quantity Ordered']
```

```python
quantity_ordered
```

```python


products=[ product for product, df in product_group]
plt.bar(products,quantity_ordered)
plt.xticks(products,rotation='vertical',size=8)
plt.show
```

```python
prices=all_data.groupby('Product').mean()['Price Each']
print (prices)
```

```python

# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
#ax1.bar(products, quantity_ordered, color='g')
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
